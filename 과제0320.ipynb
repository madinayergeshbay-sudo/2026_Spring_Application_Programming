```python
!pip install openml

import openml
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기 (OpenML에서 직접)
dataset = openml.datasets.get_dataset(31)
X, y, categorical_indicator, attribute_names = dataset.get_data(target=dataset.default_target_attribute)

df = pd.concat([X, y], axis=1)

df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
# 수치형 변수
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# 범주형 변수
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    plt.figure(figsize=(6,3))
    sns.countplot(x=col, data=df)
    plt.xticks(rotation=45)
    plt.title(f'Count of {col}')
    plt.show()
```

```python
# 상관관계
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# scatter plot
sns.scatterplot(x='age', y='credit_amount', hue='class', data=df)
plt.show()
```

```python
# boxplot
sns.boxplot(x='class', y='age', data=df)
plt.show()
```

본 연구는 독일 신용 데이터셋(credit-g)을 활용하여 고객의 신용 등급에 영향을 미치는 주요 요인을 탐색적으로 분석하는 것을 목적으로 한다.
특히 다양한 변수 간 관계를 시각화하여 신용 위험도를 이해하고, 향후 예측 모델 개발을 위한 기초 자료를 제공하고자 한다.

본 데이터는 다양한 수치형 및 범주형 변수로 구성되어 있으며, 고객의 금융 정보와 개인 특성을 포함하고 있다.
데이터 탐색 결과 결측치는 거의 존재하지 않았으며, 전체적으로 정제된 상태를 유지하고 있다.
기초 통계 분석을 통해 변수들의 분포 및 범위를 확인하였다.

수치형 변수의 경우 대부분 특정 구간에 집중된 분포를 보이며, 일부 변수에서는 이상치가 관찰되었다.
범주형 변수는 특정 클래스에 데이터가 집중되는 경향을 보였으며, 이는 데이터 불균형 가능성을 시사한다.

상관관계 분석 결과 일부 변수 간 약한 상관관계가 확인되었으며, 강한 선형 관계는 제한적으로 나타났다.
산점도 분석을 통해 변수 간 관계를 시각적으로 확인하였으며, 특정 변수 조합에서 패턴이 나타나는 것을 확인하였다.
박스플롯 분석 결과 클래스에 따라 변수 분포의 차이가 존재함을 확인하였다.

연령(age)과 신용 금액(credit_amount)은 신용 등급(class)에 영향을 미치는 중요한 변수로 판단된다.
특정 범주의 고객이 상대적으로 높은 신용 위험을 가지는 경향이 관찰되었다.
데이터 전반에서 변수 간 복합적인 관계가 존재함을 확인하였다.

본 분석을 통해 신용 등급에 영향을 미치는 주요 변수와 패턴을 파악할 수 있었다.
